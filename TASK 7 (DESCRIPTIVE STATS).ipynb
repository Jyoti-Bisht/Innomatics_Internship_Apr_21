># MEAN , MEDIAN , MODE

>>**Mean** is defined as the sum of observation divided by number of observation.It is highly affected by outliers.
   Mathematically,
                $$Mean=\Sigma_{i=0}^{n}\,\frac{x_i}{n}$$
                
>>**Median** gives you the central value of data .It splits the data into two equal halves. If the no. of observations in the      data is odd ,after sorting the array the central value is considered as the median but if it's odd,the average of 2 central      terms is taken.It has no effect of outliers.
   Mathematically,
                $$ median (provided-n-is-odd)=\frac{n+1}{2}$$
                Also,
                $$ median (provided-n-is-odd)=avg(\frac{n}{2},\frac{n}{2}+1)$$
                
>>**Mode** is the most frequently occuring value.Mainly used for categorical data and have no effect of outliers.                
                
                

# Using Formula-Mean,Median,mode

```python
import pandas as pd

x=pd.read_csv("C:/Users/Yug/Downloads/data.csv")
y=x["Annual_HH_Income"]

#Mean
mean=sum(y)//len(y)
print(mean)



#Median
n=len(y)
z=sorted(y)
if n%2!=0:
    median=z[n+1//2]
else:
    median1=z[n//2]
    median2=z[((n//2)+1)]
    median=(median1+median2)//2
print(median)    
    
    
#mode

from collections import Counter

def mod(s):
    a=Counter(s)
    return([k for k,v in a.items() if v==a.most_common(1)[0][1]])

for i in mod(y):
    print(i)


```

>## MEASURE OF VARIABILITY
>> It tells us how well our data is spreaded or distributed aound the mean.


>![pic](a.png)

>>  **Range** measures the variability between highest and lowest value.  
    **RANGE = MAXIMUM VALUE - MINIMUM VALUE**

>>   **Variance** It is defined as an average squared deviation from the mean. It is being calculated by finding the difference        between every data point and the average which is also known as the mean, squaring them, adding all of them and then            dividing by the number of data points present in our data set.

      
      
>>   **Standard Deviation** It is defined as the square root of the variance. It is being calculated by finding the Mean, then          subtract each number from the Mean which is also known as average and square the result. Adding all the values and then          divide by the no of terms followed the square root.

       

# Formula based- Variance,Std

```python
import pandas as pd
a=pd.read_csv("C:/Users/Yug/Downloads/data.csv")
x=a["Annual_HH_Income"]
n=len(x)
mean=sum(x)//n
va= sum([(x[i]-mean)**2 for i in range(n)])
var=va//n
print(var)
std=(var**(0.5))
print(std)

```

# Using Formula - Covariance

```python
import pandas as pd
import statistics
a=pd.read_csv("C:/Users/Yug/Downloads/data.csv")
b=a["Annual_HH_Income"]
bb=a["No_of_Earning_Members"]
c=statistics.stdev(b)
cc=statistics.stdev(bb)


mean1=sum(b)//len(b)
mean2=sum(bb)//len(bb)

cov=sum([((b[i]-mean1)*(bb[i]-mean2)) for i in range(len(b))])
covariance=cov//len(b)

cor=covariance//(c*cc)
print(covariance)
print(cor)
```

# Checking Skewness and Kurtosis
Skewness
In statistics, skewness is a measure of the asymmetry of the probability distribution of a random variable about its mean. In other words, skewness tells you the amount and direction of skew (departure from horizontal symmetry). The skewness value can be positive or negative, or even undefined. If skewness is 0, the data are perfectly symmetrical, although it is quite unlikely for real-world data. As a general rule of thumb:

If skewness is less than -1 or greater than 1, the distribution is highly skewed.
If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.
Kurtosis
Kurtosis:
In probability theory and statistics, kurtosis (from Greek: κυρτός, kyrtos or kurtos, meaning "curved, arching") is a measure of the "tailedness" of the probability distribution of a real-valued random variable. Like skewness, kurtosis describes the shape of a probability distribution and there are different ways of quantifying it for a theoretical distribution and corresponding ways of estimating it from a sample from a population. Different measures of kurtosis may have different interpretations.

The standard measure of a distribution's kurtosis, originating with Karl Pearson, is a scaled version of the fourth moment of the distribution. This number is related to the tails of the distribution, not its peak;hence, the sometimes-seen characterization of kurtosis as "peakedness" is incorrect. For this measure, higher kurtosis corresponds to greater extremity of deviations (or outliers), and not the configuration of data near the mean.

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
from scipy.stats import norm
for i in a.columns:
     print(i)
     print('-'*10)
     print('Skewness : ' + str(a[i].skew()))
     print('Kurtosis : '+str(a[i].kurt()))
     print('Mean : '+ str(a[i].mean()))
     print('Median : '+ str(a[i].median()))
     sns.distplot(a[i], fit=norm)
     plt.show()
     print('\n')
```
